## showing how the features works


First, read the logs of man and bot to compare them.

```python
from parser.reader import read

path_human = 'get your path'
path_bot = 'get your path'

df_human = read(path_human)
df_bot = read(path_bot)

# for more convenient comparison we will equal the sizes of logs
min_size = min(*map(len, [df_bot, df_human]))

df_human = df_human[: min_size]
df_bot = df_bot[: min_size]
```

now split into the windows

```python
from parser.windows import get_windows

human = get_windows(df_human, interval=10000)
bot = get_windows(df_bot, interval=10000)
```

in turn visualize each feature

```python
# apm 

import matplotlib.pyplot as plt
from features.features import apm

% matplotlib inline

y_human = [apm(window) for window in human]
y_bot = [apm(window) for window in bot]

plt.plot(range(len(y_human)), y_human, color='blue', label='human')
plt.plot(range(len(y_bot)), y_bot, color='red', label='bot')
plt.legend(loc='best')
plt.ylabel('APM')
plt.xlabel('TIME')
plt.show()
```
